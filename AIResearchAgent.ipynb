```python
!pip install google-genai -q

from google import genai
from google.colab import userdata # Import utility to access Colab secrets

# 1. Load the API Key from Colab Secrets
try:
    GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
    if not GEMINI_API_KEY:
        raise ValueError("GEMINI_API_KEY not found in Colab Secrets.")
except Exception as e:
    print(f"Error loading API Key: {e}")
    # Exit or handle the error gracefully
    raise SystemExit("Please set the GEMINI_API_KEY in Colab Secrets.")

# 2. Initialize the client
# The client is now defined using the explicitly loaded API key.
client = genai.Client(api_key=GEMINI_API_KEY)

# Define the model to use
model_name = "gemini-2.5-flash"

# --- Main Logic Starts Here ---

# Step 1: Get user question
topic = input("Enter your research topic: ")

# Step 2: Generate sub-questions
sub_qs_prompt = f"Generate 3 detailed research sub-questions based on the topic: {topic}"
print("--- Generating Sub-questions ---")

sub_questions_response = client.models.generate_content(
    model=model_name,
    contents=sub_qs_prompt
)

sub_questions_text = sub_questions_response.text
print("\nGenerated Sub-questions:\n", sub_questions_text)

# Step 3: Summarize into one paragraph
summary_prompt = f"Summarize the following sub-questions into one cohesive paragraph:\n{sub_questions_text}"
print("\n--- Generating Summary ---")

summary_response = client.models.generate_content(
    model=model_name,
    contents=summary_prompt
)

print("\nSummary:\n", summary_response.text)
```

```python
# --- Step 1: Get user topic ---
topic = input("Enter your research topic: ")

# --- Step 2: Generate sub-questions (rule-based templates) ---
sub_questions = [
    f"What are the main ways {topic} is currently being applied or used?",
    f"What challenges, limitations, or ethical concerns arise when considering {topic}?",
    f"How might {topic} influence future developments or change existing practices?"
]

print("\nGenerated Sub-questions:\n")
for i, q in enumerate(sub_questions, start=1):
    print(f"{i}. {q}")

# --- Step 3: Summarize into one paragraph ---
summary = (
    f"{topic} involves understanding how it is applied, the challenges or ethical "
    f"issues it presents, and the ways it may shape future developments. "
    f"Together, these aspects offer a comprehensive view of the broader impact of {topic}."
)

print("\nSummary:\n")
print(summary)
```
